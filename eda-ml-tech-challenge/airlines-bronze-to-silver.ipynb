## Importing Libraries

```python
from pyspark.sql import functions as F
import matplotlib.pyplot as plt
```

## Getting the Data

```python
airlines_file = '/Volumes/workspace/flights/bronze/flights/airlines.csv'

file_type = 'csv'

infer_schema = 'true'
first_row_is_header = 'true'
delimiter = ','
```

## Exploration (EDA)

```python
airlines = spark.read.format(file_type)\
    .option('inferSchema', infer_schema)\
        .option('header', first_row_is_header)\
            .option('sep', delimiter)\
                .load(airlines_file)
```

```python
display(airlines)
```

```python
airlines.printSchema()
```

```python
to_silver = '/Volumes/workspace/flights/silver/airlines/'

airlines.coalesce(1).write.csv(to_silver, mode='overwrite', header=True)
```
